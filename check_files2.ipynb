```python
cameras = []
dates = []
times = []
positions = []
RAs = []
DECs = []

l = 0
for i in range(all_files_number):
    if all_files_22May[i] == '\n':
        l += 1
        print ("line break", l)
    if 'sci' in all_files_22May[i]:    
        camera = all_files_22May[i][52:54]
        camera = int(camera)
        date = str(all_files_22May[i])[55:63]
        time = str(all_files_22May[i])[64:70]
        position = str(all_files_22May[i])[81:87]
        RA = int(position[0:3])
        DEC = int(position[3:7])
        cameras.append(camera)
        dates.append(date)
        times.append(time)
        positions.append(position)
        RAs.append(RA)
        DECs.append(DEC)
```

```python
file_number = len(cameras)
```
